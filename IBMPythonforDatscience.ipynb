```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go

# 1. Download Tesla stock data
tesla_data = yf.download("TSLA", start="2015-01-01", end="2023-12-31")
tesla_data.reset_index(inplace=True)

# 2. Tesla revenue data (manually created due to broken URLs)
tesla_revenue = pd.DataFrame({
    'Date': ['2023-12-31', '2023-09-30', '2023-06-30', '2023-03-31', '2022-12-31'],
    'Revenue': ['25.17B', '23.35B', '24.93B', '23.33B', '24.32B']
})

# 3. Clean revenue data
tesla_revenue['Revenue'] = tesla_revenue['Revenue'].str.replace('B', '').astype(float) * 1e9

# 4. Plotting function
def make_graph(stock_data, revenue_data, stock, title):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=stock_data['Date'], y=stock_data['Close'], name='Stock Price'))
    fig.add_trace(go.Scatter(x=revenue_data['Date'], y=revenue_data['Revenue'], name='Revenue'))
    fig.update_layout(title=title, xaxis_title='Date', yaxis_title='USD')
    fig.show()

# 5. Plot Tesla graph
make_graph(tesla_data, tesla_revenue, 'TSLA', 'Tesla Stock Price vs Revenue')
```

```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go

# 1. Download GameStop stock data
gme_data = yf.download("GME", start="2015-01-01", end="2023-12-31")
gme_data.reset_index(inplace=True)

# 2. GameStop revenue data (manually added recent values)
gme_revenue = pd.DataFrame({
    'Date': ['2023-12-31', '2023-10-31', '2023-07-31', '2023-04-30', '2023-01-31'],
    'Revenue': ['1.79B', '1.16B', '1.16B', '1.24B', '2.23B']
})

# 3. Clean revenue data
gme_revenue['Revenue'] = gme_revenue['Revenue'].str.replace('B', '').astype(float) * 1e9

# 4. Plotting function
def make_graph(stock_data, revenue_data, stock, title):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=stock_data['Date'], y=stock_data['Close'], name='Stock Price'))
    fig.add_trace(go.Scatter(x=revenue_data['Date'], y=revenue_data['Revenue'], name='Revenue'))
    fig.update_layout(title=title, xaxis_title='Date', yaxis_title='USD')
    fig.show()

# 6. Plot GameStop graph
make_graph(gme_data, gme_revenue, 'GME', 'GameStop Stock Price vs Revenue')
```
